```python
import glob
from tensorflow.python.summary.summary_iterator import summary_iterator
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
sns.set()
```

```python
def get_section_results(file):
    X = []
    Y = []
    for e in summary_iterator(file):
        for v in e.summary.value:
            if v.tag == 'Train_EnvstepsSoFar':
                X.append(v.simple_value)
            elif v.tag == 'Eval_AverageReturn':
                Y.append(v.simple_value)
    return X, Y
```

```python
logdir = 'data/ac_todo_16-12-2020_00-28-43/events.out.tfevents.1608107323.hayashintansMBP.lan'
eventfile = glob.glob(logdir)[0]
X_1, Y_1 = get_section_results(eventfile)

logdir = 'data/pg_todo_16-12-2020_00-29-34/events.out.tfevents.1608107374.hayashintansMBP.lan'
eventfile = glob.glob(logdir)[0]
X_2, Y_2 = get_section_results(eventfile)

logdir = 'data/pg_b_rtg_sa_l_4_16-12-2020_07-31-49/events.out.tfevents.1608132709.hayashintansMBP.lan'
eventfile = glob.glob(logdir)[0]
X_3, Y_3 = get_section_results(eventfile)
```

```python
plt.figure(figsize=(10,4))
plt.plot(X_1,Y_1,c='#8AC6D1',label='l=2')
plt.plot(X_2,Y_2,c='#FF8080',label='l=3')
# plt.plot(X_3,Y_3,c='#FFBA92',label='l=4')
plt.legend(bbox_to_anchor=(1.05, 0.75))
```
