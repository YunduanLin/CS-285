```python
import glob
from tensorflow.python.summary.summary_iterator import summary_iterator
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
sns.set()
```

```python
def get_section_results(file):
    X = []
    Y = []
    for e in summary_iterator(file):
        for v in e.summary.value:
            if v.tag == 'Train_EnvstepsSoFar':
                X.append(v.simple_value)
            elif v.tag == 'Eval_AverageReturn':
                Y.append(v.simple_value)
    return X, Y
```

```python
logdir = 'data/todo_15-12-2020_16-34-33/events.out.tfevents.1608078873.hayashintansMBP.lan'
eventfile = glob.glob(logdir)[0]
X_1, Y_1 = get_section_results(eventfile)

logdir = 'data/ac_todo_15-12-2020_22-38-48/events.out.tfevents.1608100728.hayashintansMBP.lan'
eventfile = glob.glob(logdir)[0]
X_2, Y_2 = get_section_results(eventfile)
```

```python
plt.figure(figsize=(10,4))
plt.plot(X_1,Y_1,c='#8AC6D1',label='PG')
plt.plot(X_2,Y_2,c='#FF8080',label='AC')
# plt.plot(X,Y_3,c='#FFBA92',label='PG-RTG-NA')
plt.legend(bbox_to_anchor=(1.05, 0.75))
```
